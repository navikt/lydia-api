```python
def nytt_tema_med_spørsmål_og_svaralternativer(tema: Tema) -> str:
    output = ""

    output += nytt_tema(tema)
    output += "\n"
    output += "\n"

    for spørsmål in tema.spørsmål:
        output += nytt_spørsmål(spørsmål)
        output += "\n"
        output += "\n"

        for svaralternativ in spørsmål.svaralternativer:
            output += nytt_svaralternativ(
                svaralternativ=svaralternativ, spørsmål=spørsmål
            )
            output += "\n"

        output += "\n"
        output += "\n"

        output += knytt_spørsmål_til_tema(spørsmål=spørsmål, tema=tema)

    return output


sql_script = ""
sql_script += gjør_tema_inaktivt(temaId=12)
sql_script += "\n"
sql_script += "\n"
sql_script += nytt_tema_med_spørsmål_og_svaralternativer(tema)

print(sql_script)
```
